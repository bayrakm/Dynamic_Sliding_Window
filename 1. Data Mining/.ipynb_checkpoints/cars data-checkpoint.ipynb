```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
conn = sqlite3.connect('cars.db')
```

```python
cur = conn.cursor()
cur.execute('SELECT * FROM cars')
```

```python
rows = cur.fetchall()
```

```python
len(rows)
```

```python
#df.to_sql('cars.db', conn)
```

```python
Date = np.array(rows[0][0])
Source =  np.array(rows[0][1])
Make =  np.array(rows[0][2])
Model =  np.array(rows[0][3])
Price=  np.array(rows[0][4])
Milage=  np.array(rows[0][5])
Year =  np.array(rows[0][6])
Power =  np.array(rows[0][7])
Consumption =  np.array(rows[0][8])
Cylinder =  np.array(rows[0][9])
Gears =  np.array(rows[0][10])
Fuel = np.array(rows[0][11])
Transmission =  np.array(rows[0][12])
Body =  np.array(rows[0][13])
CO2 = np.array(rows[0][14])
Door =  np.array(rows[0][15])
Seat =  np.array(rows[0][16])
Engine =  np.array(rows[0][17])

for row in rows:
    
    Date = np.append(Date, row[0])
    Source = np.append(Source, row[1])
    Make = np.append(Make, row[2])
    Model = np.append(Model, row[3])
    Price= np.append(Price, row[4])
    Milage= np.append(Milage, row[5])
    Year = np.append(Year, row[6])
    Power = np.append(Power, row[7])
    Consumption = np.append(Consumption, row[8])
    Cylinder = np.append(Cylinder, row[9])
    Gears = np.append(Gears, row[10])
    Fuel = np.append(Fuel, row[11])
    Transmission = np.append(Transmission, row[12])
    Body = np.append(Body, row[13])
    CO2 = np.append(CO2, row[14])
    Door = np.append(Door, row[15])
    Seat = np.append(Seat, row[16])
    Engine = np.append(Engine, row[17])
```

```python
d = {'Date': Date, 'Source':Source, 'Make':Make, 'Model':Model, 'Price':Price, 
     'Milage':Milage, 'Year':Year, 'Power':Power, 'Consumption':Consumption,
     'Cylinder':Cylinder, 'Gears':Gears, 'Fuel Type':Fuel, 'Transmission':Transmission, 
     'Body type':Body, 'CO2 emission':CO2, 'Door':Door, 'Seat':Seat, 'Engine size':Engine}
```

```python
for k,v in d.items():
    print(f'{k} : {len(v)}')
```

```python
df = pd.DataFrame(d)
```

```python
df.drop_duplicates(inplace=True)
```
