### Model

Model names contain some information we extract if we dont have it and create a concise, common model name for each make model.

```python
df.groupby(['data_source', 'Make', 'Model']).Date.count()
```

```python
df.Model.value_counts(dropna=False)
```

**Toyota**: Aygo, Auris, Yaris, RAV4, Avensis

**Vauxhall**: Astra, Corsa, ADAM, Mokka X, Insignia

**Volkswagen**: Polo, Golf, Scirocco, Tiguan, Passat,

**Ford**: Ka, Fiesta, Focus, Kuga, Mondeo

```python
def get_model(x):
    
    ''' to get model as expected
    
    x: model string
    
    return: corrected model string
    '''
    
    x = x.lower()
    
    if 'aygo' in x:
        return 'Aygo'
    elif 'auris' in x:
        return 'Auris'
    elif 'yaris' in x:
        return 'Yaris'
    elif ('rav4' in x) | ('rav 4' in x) | ('rav-4' in x):
        return 'RAV4'
    elif 'avensis' in x:
        return 'Avensis'
    elif 'astra' in x:
        return 'Astra'
    elif 'corsa' in x:
        return 'Corsa'
    elif 'adam' in x:
        return 'ADAM'
    elif ('mokka x' in x) | ('mokka' in x):
        return 'Mokka'
    elif 'insignia' in x:
        return 'Insignia'
    elif 'polo' in x:
        return 'Polo'
    elif 'golf' in x:
        return 'Golf'
    elif 'scirocco' in x:
        return 'Scirocco'
    elif 'tiguan' in x:
        return 'Tiguan'
    elif 'passat' in x:
        return 'Passat'
    elif 'ka' in x:
        return 'KA'
    elif 'fiesta' in x:
        return 'Fiesta'
    elif 'focus' in x:
        return 'Focus'
    elif 'kuga' in x:
        return 'Kuga'
    elif 'mondeo' in x:
        return 'Mondeo'
    else:
        return 'other'
```

**Outliers within Model, Year group**

```python
model = df.Model_1.unique()
year = df.Year.unique()
```

```python
def outliers_iqr(data):
    '''to get index of data points from out off 1.5 iqr range
    Parameters: data: dataframe
    
    Return: list of index of outliers
    '''
    
    idx_outliers = []
    data = data.sort_values()
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    
    # print(q1, q3)
    
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    
    # print(lwr_bound, upr_bound)
    
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            idx = data[data==i].index[0]
            idx_outliers.append(idx)
    return list(idx_outliers)
```

```python
import itertools

def index_outliers(df, col1, col2):
    
    '''to get outliers by given data and columns
    
    Parameters: df: dataframe
                col1: grouping col 1
                col2: grouping col 2
                
    Returns: list of unique index numbers of outliers
    '''
    
    idx_outliers = list()
    for i in itertools.product(model, year, repeat=1):
        data = df[(df[col1] == i[0])&(df[col2] == i[1])].Price

        if data.shape[0]>0:
            val = outliers_iqr(data)
            if val != []:
                for k in val:
                    idx_outliers.append(k)
    return list(set(idx_outliers))
```

### Divide data for Training purposes

Dividing data into chunks is important to test our project. I will use stratified chunks for sliding window approach.

```python
def data_stratification(df_, grp_col, n_grp):
    
    '''to divide data in stratified chunks.
    Remaining data to put very last chunk.
    
    Parameters: df_:dataframe
                grp_col: field to get proportions
                n_grp: number of desired group
            
    Return: the index numbers of each group within a dictionary
    
    '''
    
    idx_dict = {}

    for i in range(n_grp):

        idx=[]  

        for grp in df_[grp_col].unique():

            data = df_[df_[grp_col]==grp]
            num_data = len(data)//(n_grp)
            pro_data = len(data)/(n_grp)
            
            if n_grp > 1:

                if num_data >= 1:
                    idx.append(list(data.sample(num_data, replace=True, random_state=0).index))

                elif (pro_data <= 1) & (pro_data > 0):
                    idx.append(list(data.sample(1, replace=True, random_state=0).index))

            elif n_grp == 1:
                idx.append(list(data.index))

        idx = set([col for row in idx for col in row])    
        idx_dict[f'group{i+1}'] = idx   
        n_grp -= 1
        df_.drop(index=idx, inplace=True)
        
    return idx_dict
```

```python
df_ = df[['Model_1', 'Year']] 
df_['Group'] = df_.Model_1 +' '+ df_.Year.astype('str')
```

```python
idx_grp = data_stratification(df_, 'Group', 18)
```
